# GSEA

```python
import gseapy as gp

# Example list of genes
gene_list = ['TP53', 'BRCA1', 'EGFR', 'MYC', 'PTEN']

# Perform simple GSEA
enr = gp.enrichr(gene_list=gene_list, 
                 gene_sets='KEGG_2021_Human',  # You can choose other gene sets
                 organism='Human', 
                 outdir='gsea_results')

# Result in df
enr.results
```
